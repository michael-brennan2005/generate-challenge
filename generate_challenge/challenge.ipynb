```python
def decode(barcode):
    DEBUG = False
    def dprint(string):
        if DEBUG:
            print(string)
    
    if len(barcode) <= 1:
        return ''
    
    blocks = []
    
    current_block_digits = []
    apply_reverse = False
    apply_encrypt = False
    apply_repeat = False
    
    for char in barcode[1::]:
        dprint(f"Blocks: {blocks}\nCurrent block digits: {current_block_digits}\n----")
        
        if char.isdigit():
            dprint("New digit, appending to current block.")
            current_block_digits.append(char)
        elif char == '!':
            dprint("Repeat encountered.")
            if len(blocks) >= 1:
                current_block_digits += blocks[-1]
        elif char == '^':
            dprint("Reverse encountered.")
            apply_reverse = True
        elif char == '%':
            dprint("Encrypt encountered.")
            apply_encrypt = True
        elif char == '#':
            dprint("Ending block.")
            
            if apply_encrypt and len(blocks) >= 1:
                dprint("Encrypting previous...")
                for i in range(0, len(blocks[-1])):
                    blocks[-1][i] = str((int(blocks[-1][i]) * 2) % 10)
                    
            if apply_reverse and len(blocks) >= 1:
                dprint("Reversing previous...")
                blocks[-1].reverse()
                
            blocks.append(current_block_digits[:])
            apply_reverse = False
            apply_encrypt = False
            apply_repeat = False
            
            current_block_digits.clear()
        else:
            continue

    final_decoded = ""
    for block in blocks:
        for char in block:
            final_decoded += char
            
    return final_decoded

decode("#12#!!%^3#")
#decode("######"),decode("#12^!%%###34^#")
```

```python
# process challenges
final_result = []
for challenge in CHALLENGES:
    final_result.append(decode(challenge))
```
